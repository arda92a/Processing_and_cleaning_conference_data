```python
import pandas as pd
from rapidfuzz import fuzz

threshold = 80

to_drop = []
dropped_rows_info = []

def is_valid_index(index, df):
    return index >= 0 and index < len(df)

for i in range(len(df)):
    for j in range(i + 1, len(df)):
        if not (is_valid_index(i, df) and is_valid_index(j, df)):
            continue
        
        try:
            name_similarity = fuzz.ratio(df.loc[i, 'Name'], df.loc[j, 'Name'])
            surname_similarity = fuzz.ratio(df.loc[i, 'Surname'], df.loc[j, 'Surname'])
            
            if name_similarity > threshold and surname_similarity > threshold:
                to_drop.append(j)
                dropped_rows_info.append({
                    'dropped_index': j,
                    'dropped_name': df.loc[j, 'Name'],
                    'dropped_surname': df.loc[j, 'Surname'],
                    'similar_to_index': i,
                    'similar_name': df.loc[i, 'Name'],
                    'similar_surname': df.loc[i, 'Surname']
                })
        except KeyError as e:
            pass

df_dropped_info = pd.DataFrame(dropped_rows_info)

df_cleaned = df.drop(to_drop).reset_index(drop=True)

```

```python
import seaborn as sns 
import matplotlib.pyplot as plt


for column in ["Gender","Title"]:
    sns.countplot(data = df,x=column,color="green")
    plt.title("Columns distribution")
    plt.show()
```

```python
plt.figure(figsize=(15, 13))
sns.countplot(data = df,orient="h",y="Country",color="violet")
plt.title("Country distribution")
```

```python
gender_att = pd.DataFrame(df.groupby("Gender")["Total_Att"].sum())

sns.barplot(data = gender_att, x="Gender",y="Total_Att",color="cornflowerblue")
plt.title("Total Attendance by gender")
plt.xlabel("Gender")
plt.ylabel("Total Attendance")
plt.show()
```

```python
plt.figure(figsize=(8, 8))
plt.pie(
    gender_att["Total_Att"], 
    labels=gender_att.index, 
    autopct='%1.1f%%', 
    startangle=140, 
    colors=sns.color_palette("pastel")
)

# Grafiğin başlığı
plt.title("Total Attendance by Gender")

# Gösterim
plt.show()
```

```python
countr_att = pd.DataFrame(df.groupby("Country")["Total_Att"].sum())
countr_att.drop("Unidentified",axis=0,inplace=True)

countr_att = countr_att.sort_values("Total_Att", ascending=False).head(15)
sns.barplot(data = countr_att, y="Country", x="Total_Att",color="cornflowerblue")

plt.title("Number of attendance from the 15 countries with the highest participation")
plt.xlabel("Total Attendance")
```

```python
plt.figure(figsize=(10, 8))

explode = [0.1 if country == "China" else 0 for country in countr_att.head(5).index]

wedges, texts, autotexts = plt.pie(
    countr_att["Total_Att"].head(5), 
    labels=countr_att.head(5).index, 
    autopct='%1.1f%%', 
    startangle=140, 
    colors=sns.color_palette("pastel"),
    explode=explode  
)

plt.legend(
    wedges, 
    [f"{country}: {total_att}" for country, total_att in zip(countr_att.head(5).index, countr_att["Total_Att"].head(5))],
    title="Total Attendance",
    loc="center left", 
    bbox_to_anchor=(1, 0.5),
    fontsize='large'
)

plt.title("Participation Distribution Among the Top 5 Countries")
plt.show()
```

```python
year_columns = ["2014","2015","2016","2017","2018","2019","2020","2021","2022","2023"]

yearly_attendance = df[year_columns].sum()

plt.figure(figsize=(10, 6))
sns.lineplot(data=yearly_attendance, marker='o', color="cornflowerblue")

plt.title("Total Attendance by Year")
plt.xlabel("Year")
plt.ylabel("Total Attendance")

plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=yearly_attendance.index, y=yearly_attendance.values, color="cornflowerblue")

plt.title("Total Attendance by Year")
plt.xlabel("Year")
plt.ylabel("Total Attendance")

plt.show()
```

```python
gender_yearly_attendance = df.groupby("Gender")[year_columns].sum().T

gender_yearly_attendance = gender_yearly_attendance.reset_index().melt(id_vars="index", var_name="Gender", value_name="Total_Attendance")
gender_yearly_attendance.rename(columns={"index": "Year"}, inplace=True)

plt.figure(figsize=(12, 6))
sns.barplot(data=gender_yearly_attendance, x="Year", y="Total_Attendance", hue="Gender", palette="pastel")

plt.title("Total Attendance by Year and Gender")
plt.xlabel("Year")
plt.ylabel("Total Attendance")

plt.show()
```
